## Create a custom date-quarter column for Analysis & Comparison

```python
df_trans['dateAdded'].dt.strftime('%Y-%m')[0]
```

```python
import datetime
import time 

datetime.datetime.strptime('2017-01-01', "%Y-%m-%d")
df_trans['dateAdded'][0].date() > datetime.datetime.strptime('2016-05-01', "%Y-%m-%d").date()
```

```python
df_trans['dateAdded'][0].date().year
```

```python
def custom_dt_sorter(row):
    if row['dateAdded'].date().year == 2015:
        return '2015_Q1'
    
    elif row['dateAdded'].date().strftime('%Y-%m') == '2016-03':
        return '2016-Q1'
    
    elif datetime.datetime.strptime('2016-05-01', "%Y-%m-%d").date() \
        < row['dateAdded'].date() < datetime.datetime.strptime('2017-01-01', "%Y-%m-%d").date() :
        return '2016-Q3'
    
    elif row['dateAdded'].date() >= datetime.datetime.strptime('2017-01-01', "%Y-%m-%d").date():
        return '2017=Q1'
        
    else:
        return '??'
    

#df['race_label'] = df.apply (lambda row: label_race(row), axis=1)
```

```python
df_trans['custom_dt_sort'] =  df_trans.apply(lambda row: custom_dt_sorter(row), axis=1)
```
