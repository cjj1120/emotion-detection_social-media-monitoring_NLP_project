### Preprocess the labels column for the model 

```python
emotions = ['anger', 'fear', 'joy', 'love', 'sadness', 'surprise']
```

```python
dataset = dataset.map(lambda x : {"labels": [x[c] for c in emotions]})
```

```python
dataset['train'][0]
```

```python
# SECOND ATTEMPT 

from transformers import TrainingArguments, Trainer

import numpy as np
from datasets import load_metric

from transformers import AutoModelForSequenceClassification
from transformers import AutoTokenizer
from transformers import TrainingArguments, Trainer





model_name = 'microsoft/xtremedistil-l6-h384-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_name)


dataset = dataset.rename_column("sentence", "text")

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=64)


cols = dataset["train"].column_names
cols.remove("labels") # removed `labels` from the `to be removed list`, meaning to keep it. 
dataset_trans = dataset.map(tokenize_function, batched=True, remove_columns=cols)


import torch
dataset_trans.set_format("torch")
dataset_trans = (dataset_trans
          .map(lambda x : {"float_labels": x["labels"].to(torch.float)}, remove_columns=["labels"])
          .rename_column("float_labels", "labels"))


model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=6, problem_type="multi_label_classification")


training_args = TrainingArguments(output_dir="/content/drive/MyDrive/10.9.22-emo_class/training-microsoft xtremedistil-l6-h384-uncased/test_trainer", \
                                  evaluation_strategy="epoch", \
                                  num_train_epochs=10, \
                                  learning_rate=5e-06 #3e-03 #3e-05 
                                  )        #(0.001) per_device_train_batch_size=16, \ 

# def compute_metrics(eval_pred):
#     metric = load_metric("accuracy")
#     logits, labels = eval_pred
#     predictions = np.argmax(logits, axis=-1)
#     return metric.compute(predictions=predictions, references=labels)

trainer = Trainer(
    model=model, args=training_args, train_dataset= dataset_trans['train'], eval_dataset= dataset_trans['test']
) #, compute_metrics=compute_metrics

```

```python
trainer.train()
```
